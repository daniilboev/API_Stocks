```python
from urllib.request import urlopen
from urllib.error import HTTPError
import json
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

api = 'xxx' #use your API key from the FMP https://site.financialmodelingprep.com/developer/docs/
```

```python
def earnings(start, end, api_key, as_reported=False):
    
    if as_reported:
        URL = (f"https://financialmodelingprep.com/api/v3/earning_calendar?from={start}&to={end}&apikey={api_key}")
             
    else:
        URL = (f"https://financialmodelingprep.com/api/v3/earning_calendar?from={start}&to={end}&apikey={api_key}")

    response = urlopen(URL)
    data = json.loads(response.read().decode("utf-8"))

    data_formatted = {}
    

    return data
```

```python
start = datetime.today() - timedelta(days=1)
start = start.strftime('%Y-%m-%d')
end = datetime.today() + timedelta(days=90)
end = end.strftime('%Y-%m-%d')
```

```python
ticker_list = ['DGX',
'AAPL',
'AMAT',
'ACN',
'TER',
'CE',
'PH',
'NOC',
'EVRG',
'ES',
'WEC',
'CHD',
'GOOGL',
'NFLX',
'FB',
'TGT',
'EBAY',
'LOW',
'BBY']
```

```python
df = earnings(start, end, api, as_reported=False)
df = pd.json_normalize(df)
```

```python
df1 = df[df['symbol'].isin(ticker_list)]
df1['revenueEstimated_bn'] = df1['revenueEstimated'] / 1000000000
df1 = df1[['date','symbol','epsEstimated','revenueEstimated_bn']]
df1
```
